# DIP Lab 6: Image Enhancement

```python
from PIL import Image, ImageOps, ImageFilter

image = Image.open('1.jpg')
image
```

### Histogram Equalization:

```python
# Histogram Equalization
image_equalized = ImageOps.equalize(image)
image_equalized
```

```python
# Plotting Histogram
import matplotlib.pyplot as plt
def histogram(image,text):
    image = image.convert('L')
    data = list(image.getdata())
    plt.hist(data, density=True,bins=255)
    plt.title(text)
    plt.show()

# Original Image Histogram
histogram(image,"Original Image")
# Equalized Image Histogram
histogram(image_equalized,"Histogram Equalized Image")
```

### Contrast Stretching:

```python
# Contrast Stretching
import numpy as np

max_value = np.amax(image)
min_value = np.amin(image)
diff = max_value - min_value
width, height = image.size

for x in range(0,width):
    for y in range(0, height):
        pixel = image.getpixel((x,y))
        pixel = (pixel - min_value) * (255 // diff)
        image.putpixel((x,y),tuple(pixel))
image
```
