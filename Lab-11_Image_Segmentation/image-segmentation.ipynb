```python
import cv2
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('Untitled.png',0)
Image.fromarray(image)
```

```python
ret2,otsu_threshold = cv2.threshold(image,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
Image.fromarray(otsu_threshold)
```

```python
kernel = np.ones((11,11),np.uint8)
dilation = cv2.dilate(otsu_threshold,kernel,iterations = 1)
kernel = np.ones((5,5),np.uint8)
closing = cv2.morphologyEx(dilation,cv2.MORPH_CLOSE,kernel)
Image.fromarray(closing)
```

```python
%matplotlib inline
ret,labels = cv2.connectedComponents(closing)
plt.imshow(labels, cmap="nipy_spectral")
```
