# DIP LAB 10: Erosion and Dilation

### Task-1:

```python
import cv2
import numpy as np
from PIL import Image

img = cv2.imread('croft.png',0)
kernel = np.ones((3,3),np.uint8)
erosion1 = cv2.erode(img,kernel,iterations = 1)
erosion2 = cv2.erode(img,kernel,iterations = 2)

dilation1 = cv2.dilate(img, kernel, iterations=1) 
dilation2 = cv2.dilate(img, kernel, iterations=2)

print('Erosion Iteration 1')
display(Image.fromarray(erosion1))
print('Erosion Iteration 2')
display(Image.fromarray(erosion2))
print('Dilation Iteration 1')
display(Image.fromarray(dilation1))
print('Dilation Iteration 2')
display(Image.fromarray(dilation2))
```
